# APIアクセス

## whether hacks

```python
api_url = 'http://weather.livedoor.com/forecast/webservice/json/v1?city=270000'
```

```python
whether_data = requests.get(api_url).json()
```

```python
pprint.pprint(whether_data)
```

```python
# パラメータを辞書型にして与える
api_url = 'http://weather.livedoor.com/forecast/webservice/json/v1'
city_code = {'city':270000}
whether_data = requests.get(api_url, params = city_code).json()
```

```python
pprint.pprint(whether_data)
```

```python
#取得データのキーを取得
whether_data.keys()
```

```python
pprint.pprint(whether_data['forecasts'])
```

```python
pprint.pprint(whether_data['forecasts'][0])
```

## MediaWiki

```python
import requests, pprint
```

```python
api_url = "https://ja.wikipedia.org/w/api.php"
api_params = {
    'format':'json', 
    'action':'query', 
    'titles':'安室奈美恵', 
    'prop':'revisions', 
    'rvprop':'content'}
wiki_data = requests.get(api_url, params = api_params).json()
```

```python
pprint.pprint(wiki_data)
```
